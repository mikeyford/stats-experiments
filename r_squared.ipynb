```
from sklearn.metrics import r2_score
from numpy import *
```

```
y = [1, 2, 4]
y_hat = [1.1, 2, 1.5]
y_bar = mean(y)

sst = sum(list(map(lambda x: (x - y_bar)**2 , y)))
ssr = sum(list(map(lambda x: x**2, (array(y)-array(y_hat)))))
print("SST:", sst)
print("SSR:", ssr)
1 - (ssr/sst)
```

```
r2_score(y, y_hat)
```

```
import pandas as pd
%matplotlib inline

pd.DataFrame({'y':y, 'y_hat':y_hat}).plot()
```
