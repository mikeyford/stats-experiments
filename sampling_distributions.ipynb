```python
import numpy as np
import seaborn as sns
import warnings
from functools import partial

sns.set(color_codes=True)
warnings.filterwarnings("ignore", category=UserWarning)
```

```python
def distribution_of_sample_stat(stat, population_dist, sample_size, samples):     
    results = [stat(population_dist(size=sample_size)) for i in range(0, samples)]    
    print("mean:", round(np.mean(results),3))
    print("standard deviation:", round(np.std(results),3))        
    return sns.distplot(results)
```

```python
distribution_of_sample_stat(np.mean, np.random.normal, 10, 10)
# distribution of sample mean - sample size of 10, with 10 samples
```

```python
distribution_of_sample_stat(np.mean, np.random.normal, 10, 100)
# distribution of sample mean - sample size of 10, with 100 samples
```

```python
distribution_of_sample_stat(np.mean, np.random.normal, 10, 1000)
# distribution of sample mean - sample size of 10, with 1000 samples
```

```python
distribution_of_sample_stat(np.mean, np.random.normal, 10, 10000)
# distribution of sample mean - sample size of 10, with 10,000 samples
```

```python
distribution_of_sample_stat(np.mean, np.random.normal, 1000, 10000)
# distribution of sample mean - sample size of 1000
```

```python
distribution_of_sample_stat(np.mean, np.random.uniform, 1000, 10000)
# distribution of sample mean from uniform distribution - sample size of 100
```

```python
distribution_of_sample_stat(partial(np.percentile, q=75), np.random.normal, 100, 10000)
# distribution of sample third quartile - sample size of 100
```

```python
distribution_of_sample_stat(partial(np.percentile, q=75), np.random.normal, 1000, 10000)
# distribution of sample third quartile - sample size of 1000
```

```python
distribution_of_sample_stat(partial(np.percentile, q=75), np.random.uniform, 1000, 10000)
# distribution of sample third quartile - sample size of 1000
```

```python
distribution_of_sample_stat(np.min, np.random.normal, 100, 10000)
# distribution of sample third quartile - sample size of 100
```

```python
distribution_of_sample_stat(np.min, np.random.normal, 1000, 10000)
# distribution of sample third quartile - sample size of 1000
```

```python
distribution_of_sample_stat(np.min, np.random.normal, 100, 1000000)
# distribution of sample third quartile - sample size of 100, 1,000,000 samples
```

```python
distribution_of_sample_stat(np.min, np.random.uniform, 100, 1000)
# distribution of sample third quartile - sample size of 100 - uniform distribution
```

```python
distribution_of_sample_stat(np.min, np.random.uniform, 100, 10000)
# distribution of sample third quartile - sample size of 100 - uniform distribution, 10,000 samples
```
